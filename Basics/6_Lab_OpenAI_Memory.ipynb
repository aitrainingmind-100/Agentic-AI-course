```python
import os
from openai import OpenAI

def main():
    print("OpenAI Interactive Chat (With Memory)")
    print("Type 'exit' to quit\n")

    client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

    # ðŸ§  Memory: Store conversation history
    messages = [
        {"role": "system", "content": "You are a helpful AI assistant."}
    ]

    while True:
        user_input = input("You: ").strip()

        if user_input.lower() in ["exit", "quit", "q"]:
            print("Goodbye!")
            break

        if not user_input:
            continue

        # Add user message to memory
        messages.append({"role": "user", "content": user_input})

        response = client.chat.completions.create(
            model="gpt-5-mini",
            messages=messages
        )

        assistant_reply = response.choices[0].message.content

        # Add assistant reply to memory
        messages.append({"role": "assistant", "content": assistant_reply})

        print("\nModel:", assistant_reply, "\n")


if __name__ == "__main__":
    main()
```

```python
import os
import json
from openai import OpenAI

MEMORY_FILE = "chat_memory.json"


def load_memory():
    if os.path.exists(MEMORY_FILE):
        with open(MEMORY_FILE, "r") as f:
            return json.load(f)
    return [{"role": "system", "content": "You are a helpful AI assistant."}]


def save_memory(messages):
    with open(MEMORY_FILE, "w") as f:
        json.dump(messages, f, indent=2)


def main():
    print("OpenAI Chat (Persistent Memory)")
    print("Type 'exit' to quit | 'clear' to reset memory\n")

    client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

    messages = load_memory()

    while True:
        user_input = input("You: ").strip()

        if user_input.lower() in ["exit", "quit", "q"]:
            save_memory(messages)
            print("Goodbye!")
            break

        if user_input.lower() == "clear":
            messages = [{"role": "system", "content": "You are a helpful AI assistant."}]
            save_memory(messages)
            print("Memory cleared.\n")
            continue

        if not user_input:
            continue

        messages.append({"role": "user", "content": user_input})

        response = client.chat.completions.create(
            model="gpt-5-mini",
            messages=messages
        )

        reply = response.choices[0].message.content

        messages.append({"role": "assistant", "content": reply})
        save_memory(messages)

        print("\nModel:", reply, "\n")


if __name__ == "__main__":
    main()
```
