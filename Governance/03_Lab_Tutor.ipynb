```python
import os
from dotenv import load_dotenv
from openai import OpenAI

load_dotenv()
client = OpenAI()

SYSTEM = """
You are a patient study tutor.

GUARDRAILS:
- Explain step-by-step.
- Use simple language.
- After explanation, ask: "Would you like a deeper explanation?"
- Never skip reasoning steps.
"""

def main():
    print("Study Tutor (type 'exit' to quit)\n")

    while True:
        question = input("Question: ").strip()
        if question.lower() == "exit":
            break

        response = client.chat.completions.create(
            model="gpt-5-mini",
            messages=[
                {"role": "system", "content": SYSTEM},
                {"role": "user", "content": question}
            ]
        )

        print("\nTutor:\n", response.choices[0].message.content, "\n")

if __name__ == "__main__":
    main()
```
