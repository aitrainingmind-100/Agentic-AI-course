```python
import psycopg2
from faker import Faker

# install dependencies if needed:
# pip install psycopg2-binary faker

fake = Faker()

conn = psycopg2.connect(
    host="test_postgres",
    database="mydb",
    user="myuser",
    password="mypassword",
    port=5432
)

cursor = conn.cursor()

for i in range(100):
    username = f"user{i+1}"
    email = f"user{i+1}@example.com"
    password_hash = f"hashed_password_{i+1}"
    full_name = fake.name()

    cursor.execute(
        """
        INSERT INTO users (username, email, password_hash, full_name)
        VALUES (%s, %s, %s, %s)
        """,
        (username, email, password_hash, full_name)
    )

conn.commit()
cursor.close()
conn.close()

print(" Inserted 100 sample users successfully!")

```
