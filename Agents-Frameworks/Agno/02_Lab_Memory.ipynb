```python
# 02_agno_memory.py
from dotenv import load_dotenv
load_dotenv()

from agno.agent import Agent
from agno.db.sqlite import SqliteDb
from agno.models.openai import OpenAIChat

def main():
    db = SqliteDb(db_file="agno_memory.db")

    agent = Agent(
        name="MemoryAgent",
        model=OpenAIChat(id="gpt-4o-mini"),
        db=db,
        update_memory_on_run=True,   # ✅ automatic memory
        markdown=True,
    )

    agent.print_response("My name is Anandh and I prefer email over phone.", markdown=True)
    agent.print_response("What’s the best way to reach me?", markdown=True)

if __name__ == "__main__":
    main()
```
