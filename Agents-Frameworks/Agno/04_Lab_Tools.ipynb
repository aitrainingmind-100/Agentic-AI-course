```python
# 04_agno_tools.py
from dotenv import load_dotenv
load_dotenv()

from agno.agent import Agent
from agno.models.openai import OpenAIChat
from agno.tools.duckduckgo import DuckDuckGoTools
from agno.tools import tool

@tool
def word_count(text: str) -> int:
    print(f"\n[TOOL CALL] word_count(text='{text}')")
    result = len([w for w in text.strip().split() if w])
    print(f"[TOOL RESULT] {result}\n")
    return result

def main():
    agent = Agent(
        name="ToolsAgent",
        model=OpenAIChat(id="gpt-4o-mini"),
        tools=[DuckDuckGoTools(), word_count],
        markdown=True,
        debug_mode=True,
    )

    print("Type a task. Type exit to quit.")

    while True:
        q = input("\nTask> ").strip()
        if not q or q.lower() in {"exit", "quit"}:
            break

        print("\n--- AGENT RUN START ---")
        agent.print_response(q, markdown=True)
        print("--- AGENT RUN END ---\n")

if __name__ == "__main__":
    main()
```
