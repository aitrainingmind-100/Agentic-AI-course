```python
# 03_agno_session_management.py
from dotenv import load_dotenv
load_dotenv()

from agno.agent import Agent
from agno.db.sqlite import SqliteDb
from agno.models.openai import OpenAIChat

def main():
    db = SqliteDb(db_file="agno_sessions.db")

    agent = Agent(
        name="SessionAgent",
        model=OpenAIChat(id="gpt-4o-mini"),
        db=db,
        add_history_to_context=True,  # âœ… include current session chat history
        num_history_runs=6,
        markdown=True,
    )

    user_id = "demo-user"
    session_id = "default"

    print("Type `session:<name>` to switch sessions (e.g., session:alice). Type exit to quit.")
    while True:
        text = input(f"\n[{session_id}] You> ").strip()
        if not text:
            continue
        if text.lower() in {"exit", "quit"}:
            break

        if text.lower().startswith("session:"):
            session_id = text.split(":", 1)[1].strip() or "default"
            print(f"Switched to session: {session_id}")
            continue

        # session isolation happens via (user_id, session_id)
        agent.print_response(
            text,
            markdown=True,
            user_id=user_id,
            session_id=session_id,
        )

if __name__ == "__main__":
    main()
```
