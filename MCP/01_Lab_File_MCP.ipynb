```python
import os
from langchain_openai import ChatOpenAI
from mcp_use import MCPAgent, MCPClient

async def main():
    config = {
        "mcpServers": {
            "filesystem": {
                "command": "npx",
                "args": [
                    "-y",
                    "@modelcontextprotocol/server-filesystem",
                    r"/Users/anandh/projects/Agentic-AI-course/Governance",
                   
                ],
            }
        }
    }

    client = MCPClient.from_dict(config)

    llm = ChatOpenAI(
        model="gpt-4o-mini",
        temperature=0.2,
        api_key=os.environ["OPENAI_API_KEY"]
    )

    agent = MCPAgent(llm=llm, client=client, max_steps=10, memory_enabled=True)

    prompt = "List files in the allowed folders and summarize what looks important."
    result = await agent.run(prompt)

    with open("output.html", "w", encoding="utf-8") as f:
        f.write(result)

    return result

# In Jupyter, run this in a cell:
result = await main()
print(result)
```
